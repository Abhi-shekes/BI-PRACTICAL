```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/weather_game.csv')
print(data)
```

```python
# Initial Data Exploration
print("First 6 rows of the dataset:")
print(data.head(6))
```

```python
# Summary Statistics
print("\nSummary Statistics:")
print(data.describe())
```

```python
# Get information about the dataset, including data types and missing values
print("\nData information:")
print(data.info())
```

```python
# Visualizing Categorical Variables
plt.figure(figsize=(8, 6))
sns.countplot(x='Outlook', data=data)
plt.title('Count of Outlook')
plt.xlabel('Outlook')
plt.ylabel('Count')
plt.show()
```

```python

plt.figure(figsize=(8, 6))
sns.countplot(x='Wind', data=data)
plt.title('Count of Wind')
plt.xlabel('Wind')
plt.ylabel('Count')
plt.show()
```

```python
# Visualizing Numerical Variables
plt.figure(figsize=(8, 6))
sns.histplot(data['Temperature'], bins=8, kde=False)
plt.title('Temperature Distribution')
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.show()
```

```python

plt.figure(figsize=(8, 6))
sns.histplot(data['Humidity'], bins=10, kde=True)
plt.title('Humidity Distribution')
plt.xlabel('Humidity')
plt.ylabel('Frequency')
plt.show()
```
